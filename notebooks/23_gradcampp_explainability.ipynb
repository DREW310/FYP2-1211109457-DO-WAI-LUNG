```python
from pathlib import Path
import csv
import json

try:
    import pandas as pd
except ImportError as error:
    raise ImportError("Please install pandas in the active environment before running this notebook.") from error

from IPython.display import display


# This helper keeps the notebook easy to run from the repo root or from notebooks/.
def find_repo_root(start_path: Path) -> Path:
    if (start_path / "raw_data").exists():
        return start_path
    if start_path.name == "notebooks" and (start_path.parent / "raw_data").exists():
        return start_path.parent
    raise FileNotFoundError("Could not find the FYP2 repo root.")

REPO_ROOT = find_repo_root(Path.cwd())
RESULTS_DIR = REPO_ROOT / "results"
NOTEBOOK_TAG = "23_gradcampp_explainability"
NOTEBOOK_RESULTS_DIR = RESULTS_DIR / NOTEBOOK_TAG
NOTEBOOK_RESULTS_DIR.mkdir(parents=True, exist_ok=True)
DIBAS_COMPARE_PATH = RESULTS_DIR / "11_compare_main_results_dibas" / "model_comparison.csv"
DATASET2_COMPARE_PATH = RESULTS_DIR / "21_compare_main_results_dataset2" / "model_comparison.csv"
```

```python
# This helper writes a CSV file with a stable header.
def write_csv_rows(csv_path: Path, rows, fieldnames):
    if not rows:
        raise ValueError(f"No rows were provided for {csv_path.name}.")
    csv_path.parent.mkdir(parents=True, exist_ok=True)
    with csv_path.open("w", newline="", encoding="utf-8") as handle:
        writer = csv.DictWriter(handle, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(rows)


# This helper writes one small JSON file with clean formatting.
def write_json(json_path: Path, data):
    json_path.parent.mkdir(parents=True, exist_ok=True)
    json_path.write_text(json.dumps(data, indent=2), encoding="utf-8")


# This helper stores paths in repo-relative form so the outputs stay portable.
def as_repo_relative(repo_root: Path, path_value: Path) -> str:
    return path_value.relative_to(repo_root).as_posix()
```

```python
dibas_ready = DIBAS_COMPARE_PATH.exists()
dataset2_ready = DATASET2_COMPARE_PATH.exists()
status_rows = [
    {"input_name": "dibas_model_comparison", "exists": int(dibas_ready), "path": as_repo_relative(REPO_ROOT, DIBAS_COMPARE_PATH)},
    {"input_name": "dataset2_model_comparison", "exists": int(dataset2_ready), "path": as_repo_relative(REPO_ROOT, DATASET2_COMPARE_PATH)},
]
write_csv_rows(NOTEBOOK_RESULTS_DIR / "input_status.csv", status_rows, list(status_rows[0].keys()))
write_json(NOTEBOOK_RESULTS_DIR / "summary.json", {"notebook_tag": NOTEBOOK_TAG, "dibas_ready": dibas_ready, "dataset2_ready": dataset2_ready})
display(pd.DataFrame(status_rows))
print("This notebook is a placeholder until the rebuilt main experiments finish.")
```
